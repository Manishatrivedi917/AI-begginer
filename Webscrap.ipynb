```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,/;q=0.8",
    "Accept-Language": "en-US,en;q=0.9"
}
wbpg=requests.get("https://www.ambitionbox.com/list-of-companies?page=1", headers=headers ).text
```

```python
soup=BeautifulSoup(wbpg,'lxml')
```

```python
print(soup.prettify())
```

```python
comp=soup.find_all('div' ,class_="companyCardWrapper")
```

```python
comp
```

```python
name=[]
rating=[]
review_count=[]
sal=[]

for i in comp:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('div',class_='rating_star_container').text.strip())
  review_count.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[0].text.strip())
  sal.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[1].text.strip())
```

```python
df={'name':name,'review':review_count,'rating':rating,'salary':sal}
```

```python
final=pd.DataFrame()
for j in range(1,15):
    url="https://www.ambitionbox.com/list-of-companies?page={}".format(j)

    headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,/;q=0.8",
    "Accept-Language": "en-US,en;q=0.9"
}
    wbpg=requests.get(url, headers=headers ).text
    soup=BeautifulSoup(wbpg,'lxml')
    comp=soup.find_all('div' ,class_="companyCardWrapper")
    name=[]
    rating=[]
    review_count=[]
    sal=[]

    for i in comp:
     name.append(i.find('h2').text.strip())
     rating.append(i.find('div',class_='rating_star_container').text.strip())
     review_count.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[0].text.strip())
     sal.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[1].text.strip())
    df={'name':name,'review':review_count,'rating':rating,'salary':sal}
    data=pd.DataFrame(df)
    final = pd.concat([final, data], ignore_index=True)        
    
```
