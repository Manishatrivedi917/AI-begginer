```python
#**VIA SICK-IT LEARN**
from sklearn.preprocessing import OneHotEncoder
```

```python
sd = OneHotEncoder(sparse_output=False)
```

```python
sd_data = sd.fit_transform(df[['Gender', 'City', 'Color']])
```

```python
sd_data
```

```python
encoded_cols = sd.get_feature_names_out(['Gender', 'City', 'Color'])
```

```python
encoded_cols
```

```python
print(sd_data.shape)
print(len(encoded_cols))
```

```python
encoded_df = pd.DataFrame(sd_data, columns=encoded_cols)

final_df = pd.concat([df[['Name']], encoded_df], axis=1)

print(final_df)
```
