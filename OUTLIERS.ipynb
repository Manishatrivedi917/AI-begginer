```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('placement.csv')
df.sample(6)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df['cgpa'], kde=True)
plt.title('Distribution of CGPA')

plt.subplot(1,2,2)
sns.histplot(df['iq'], kde=True)
plt.title('Distribution of IQ')

plt.tight_layout()
plt.show()
```

```python
print('Mean of IQ',df['iq'].mean())
print('St.Deviation ',df['iq'].std())
print('Min',df['iq'].min())
print('Max',df['iq'].max())
```

```python
print('Highest allowed ', df['iq'].mean()+ 3*df['iq'].std())
print('Lowst allowed ', df['iq'].mean()-3*df['iq'].std())
```
